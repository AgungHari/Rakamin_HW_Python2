```python
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD',
                   'B 1111 AM', 'B 6726 D', 'D 11223 KJ', 'AE 44677 GH', 'AE 67269 AA']

plat_genap = []
plat_ganjil = []

for plat in list_plat_mobil:
    angka = int(plat.split()[1])
    if angka % 2 == 0:
        plat_genap.append(plat)
    else:
        plat_ganjil.append(plat)

print("Plat yang genap adalah:", plat_genap)
print("Plat yang ganjil adalah:", plat_ganjil)
print("Jumlah plat genap:", len(plat_genap))
print("Jumlah plat ganjil:", len(plat_ganjil))
```

# 3 
Team Campaign PT ABC akan membuat promo bundling category product (berisi 3
category setiap bundling) untuk promo 17 Agustus. PT ABC memiliki 10 category product.

Ketentuan Bundling:
 -Tidak boleh hanya memiliki 1 jenis category dalam 1 bundling
 
 -1 Bundling boleh memiliki 2 category yang sama
 
 -Urutan tidak menjadi faktor pembeda (c1 - c2 - c3 = c2 - c1 - c3).

Tentukan ada berapa jumlah kemungkinan bundling yang bisa terbentuk?

Berikut list code category list_id_category = [c1, c2, c3, c4, c5, c6, c7 , c8, c9, c10] (10 poin)

Langkah-langkah:

a. Buatlah variable kosong untuk menampung kemungkinan kombinasi

b. Gunakan penerapan nested for (for dalam for)

c. Buatlah logic pengecekan apakah kombinasi 3 categori sudah ada pada variable inisiasi awal

d. Manfaatkan fungsi sort untuk mengurutkan urutan kombinasi

Untuk mempermudah kamu, yuk lihat resource di bawah ini:

● Sesi 2 Iterasi bagian nested for dan Solusi Soal 3 Exercise pada LMS

```python
list_id_category = ['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 'c10']

bundling_list = []

for x in list_id_category:
    for y in list_id_category:
        for z in list_id_category:
            if x != y and x != z and y != z:
                bundling = sorted([x, y, z])            # urutkan bundling
                bundling_str = '-'.join(bundling)       
                if bundling_str not in bundling_list:   # mencegah agar tidak ada kombinasi yg sama
                    bundling_list.append(bundling_str)  

print(bundling_list)
print('Jumlah probabilitas bundling:', len(bundling_list))
```

# 4
Buatlah sebuah function string_processor dengan inputan 2 parameter, yaitu suatu

string 1 kata dan nama proses yang akan dilakukan. Function ini memiliki 4 proses

a. first_char : return huruf pertama dari string yang diinput

b. last_char : return huruf terakhir dari string yang diinput

c. num_chars : return jumlah huruf string input

d. num_vowels : return jumlah huruf vocal pada string input

(12 poin)

Langkah-langkah:

a. Buatlah function baru bernama string_processor

b. Function ini memiliki 2 parameter

c. Buatlah conditional statement di dalam function nya sesuai dengan opsi proses yang dipilih ketika melakukan function call

Untuk mempermudah kamu, yuk lihat resource di bawah ini:

● Sesi 3 pada Materi Function

```python
def string_processor(x, metode = 'first_char'):

    if metode == 'first_char':
        result = x[0]
    elif metode == 'last_char':
        result = x[-1]
    elif metode == 'num_chars':
        result = len(x)
    elif metode == 'num_vowels':
        result = sum(1 for i in x.lower() if i in 'aiueo')  # hitung vokal
    else:
        result = 'Inputan Error'
    return result
```

```python
print(string_processor('martabak','first_char'))
print(string_processor('martabak','last_char'))
print(string_processor('martabak','num_chars'))
print(string_processor('martabak','num_vowels'))
print(string_processor('martabak','num_vowelssss'))
```

Menggunakan dataset superstore! Buatlah function yang bisa menghitung Jumlah Sales
dan Profit pada suatu tahun dan region tertentu dengan default tahun 2017.

Tentukan total sales dan profit:

 1.South Region pada tahun 2016

 2.West Region pada tahun 2017

(14 poin)

Langkah-langkah:
a. Buatlah function dengan 2 parameter, salah satu parameter nya merupakan default value 2017

b. Gunakan column Sales dan Profit sebagai metric yang di agregasi (group by)

Untuk mempermudah kamu, yuk lihat resource di bawah ini:

```python
!pip install xlrd
```

```python
df1 = pd.read_excel('superstore.xls')
```

```python
df1['year'] = pd.DatetimeIndex(df1['Order Date']).year
```

```python
df1.info()
```

```python
df1.head()
```

```python
def hitung_sales_profit(df1, region, year=2017):
    if year:
        df_filtered = df1[(df1['Region'] == region) & (df1['year'] == year)]
    else:
        df_filtered = df1[df1['Region'] == region]
    
    total_sales = df_filtered['Sales'].sum()
    total_profit = df_filtered['Profit'].sum()
    
    return total_sales, total_profit
```

```python
# 1. South Region tahun 2016
sales_2016_south, profit_2016_south = hitung_sales_profit(df1, "South", 2016)

# 2. West Region tahun 2017 (default)
sales_2017_west, profit_2017_west = hitung_sales_profit(df1, "West")

print('South - 2016 dengan total Sales', sales_2016_south,'dan total Profit', profit_2016_south)
print('West - 2017 dengan total Sales:', sales_2017_west,'dan total Profit', profit_2017_west)
```

```python
# Agregasi total sales per customer per tahun
df_agg = df1.groupby(['Customer ID', 'year'])['Sales'].sum().reset_index()

# Cek statistik deskriptif
print(df_agg['Sales'].describe())

def segmentasi_sales(sales):
    try:
        if sales >= 1172:
            return 'High'
        elif sales >= 121:
            return 'Medium'
        else:
            return 'Low'
    except:
        return 'Unknown'

df_agg['Sales Segment'] = df_agg['Sales'].apply(segmentasi_sales)
```

```python
df_2017 = df_agg[df_agg['year'] == 2017]

hasil = df_2017.groupby('Sales Segment')['Customer ID'].nunique().reset_index()
hasil.columns = ['Sales Segment', 'Jumlah Customer']

print(hasil)
```
